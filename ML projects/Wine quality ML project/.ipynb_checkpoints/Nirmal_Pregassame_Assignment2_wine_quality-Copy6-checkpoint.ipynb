# Q2: Implication

Demonstrating implication writen in Python code 


```python
good = df[df.grade == 1]
notgood = df[df.grade == 0]
```

```python
drop_items = ['quality','grade']
g1 = pd.DataFrame(good.drop(drop_items, axis=1).mean(), columns=['Good']).T
g2 = pd.DataFrame(notgood.drop(drop_items, axis=1).mean(), columns=['Not Good']).T
total = pd.DataFrame(df.drop(drop_items, axis=1).mean(), columns=['Total Average']).T
data = g1.append([g2, total])
```

```python
# Set standard
temp1 = data.values.reshape((3, 11))
standard = data.loc['Total Average'].values.reshape((1, 11))
temp = 100* temp1 / standard
data_percentage = pd.DataFrame(temp, columns = data.columns.values.tolist())
```

```python
from math import pi
Attributes =list(data_percentage)
AttNo = len(Attributes)

values = data_percentage.iloc[0].tolist() #
values += values [:1]

angles = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles += angles [:1]

values2 = data_percentage.iloc[1].tolist() # 
values2 += values2 [:1]

angles2 = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles2 += angles2 [:1]

values3 = data_percentage.iloc[2].tolist() # 
values3 += values3 [:1]

angles3 = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles3 += angles3 [:1]

plt.figure(figsize=(10,10))
ax = plt.subplot(111, polar=True)
plt.xticks(angles[:-1],Attributes)

# Good 
ax.plot(angles, values, color = 'r')
ax.fill(angles, values, 'red', alpha=0.1)

# Not Good
ax.plot(angles2, values2, color = 'b')
ax.fill(angles2, values2, 'blue', alpha=0.1)

# Total Average
ax.plot(angles3, values3, color = 'black')
ax.fill(angles3, values3, 'black', alpha=0.1)

plt.figtext(0.4,0.82,'Good Quality Average',color='red')
plt.figtext(0.28,0.48,'Not Good Quality Average',color='blue')
plt.figtext(0.36,0.23,'Total Average',color='black')
plt.show()
```

# Q4
The operation of Forward chaining consist on :
1) Fire/remove any rule whose premises are satisfied in the
Knowledge Base
2) Add those rules to the Knowledge Base
3) Then  Repeat the above 2 operations until a conclusion can be given on the final sentence

We have though to assume that the KB is of Horn clauses.
The forward chaining is data driven


The operation of Backward chaining is the reverse operation in comparison to forward chaining.
The final atomic sentence is known and proven and the operation consiste on recursively search for the premises which proves the final sentence.
The backward chaining is goal driven.
and just like the forward chaining, we have to assume that the KB is of Horn clauses.

The main disadvantages of Forward chaining is that at a given step, the choice between the firable rules is random. Therefore, it may apply rules that are useless for the proof of the final sentence.

The main disadvantage of backward chaining is that it follows a particular path of reasoning even if it proves that it may be pointless.
