   ### 2.1.3 Scaling data
    
    A quick look to the dataset table displayed in the section 2.1.2 will show that each feature has a different scale.
    Typically features such as volatile acidity, citric acid, chlorides, density or sulphates have values which never exceed 1 whereas the total sulfure dioxyde parameter can have values above 100.
    The purpose of scaling the data is precisely to avoid the large inequality of scale from one feature value to another.
    In reality, the is 2 ways in dealing with dataset features with different ranges: by normalizing the data or standardizing the data.
    Normalizing the data scales all features values from 0 to 1
    Standardizing the data will keep the data within a normal distribution, with a mean of 0 and a standard deviation of 1.
    The normalizing proceess is especially good for all Machine Learning model which needs to keep the distance ( the standardization does not keep the distance scale properly) whereas the standardization is more efficient for clustering technique.
    For our futur study, we will apply the normalization technique to knn model and the standardization technique for the other model.
    

```python
from sklearn import preprocessing

# Using the preprocessing Normalization to scale the data
scalingObj = preprocessing.MinMaxScaler()
newdf = df.copy()

# We normalize features data  only not the target criteria
newdf[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']] = scalingObj.fit_transform(newdf[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']])

print('Normalized data')
print(newdf)
```

   ### 2.1.4 Imbalanced data
    
    Checking imbalanced data consists in checking the distribution of target result between each result categories. A proper balanced data have the target equally repartited between each category .
    The risk for an imbalanced data is that the Machine Model will be specifically trained for the majority class (data which result is more represented). Therefore, we may have a good accuracy only for the majority class and not the class under represented.
    In order to check if the dataset is imbalanced, we need to see the repartition of target result. A pie chart can help us for that. 
    

```python
import matplotlib.pyplot as plt

# We will construct the pie chart to visualize the repartition of wine quality
plt.figure(figsize = (8,8))
labels = df.quality.value_counts().index
plt.pie(df.quality.value_counts(), autopct='%1.1f%%')
plt.legend(labels, loc="best")
plt.axis('equal')
plt.title('Quality Pie Chart')
plt.show()
```

    
    The pie chart reveals that each category of the wine quality is imbalanced.
    they are few wine in quality category 3, 4 , 7 and 8.
    Most of the wines are in quality category 6 and 5.
    In this context, it is quite difficult to exploit the data as it is.
    A possibility is to divide the quality into 2 grades (categories) :'good wine' (grade 1) and 'bad wine' (grade 0).
    We can for example set the 'bad wine' grade for wines below the quality of 7 and the rest as 'good wine'.  


```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# We create a new target criteria: the grade to check if the wine is good or bad
df['grade'] = 1 # good
df.grade[df.quality < 7] = 0 # not good

# We will construct the pie chart to visualize the repartition of wine grade
plt.figure(figsize = (8,8))
labels = df.grade.value_counts().index
plt.pie(df.grade.value_counts(), autopct='%1.1f%%')
plt.legend(labels, loc="best")
plt.axis('equal')
plt.title('Quality Pie Chart')
plt.show()
print('The good quality wines count for ',round(df.grade.value_counts(normalize=True)[1]*100,1),'%.')
```

    
    
    By doing so, the data is a little bit more exploitable as we avoid very tiny categories . But we can notice that the dataset is still imbalanced as we have more 'bad wine' than 'good wine'.
    We have therefore to make another process in order to make the data more balanced.
    An oversampling technique like SMOTE can indeed, create new artificial data points for the minority class in our dataset and therefore rebalance the data.
    By rebalancing the data, we can avoid the ML to be trained exclusively on the majority class and consequently, we can improve the confusion matrix.
    You can find below the SMOTE technique applied in our dataset and its impact on the confusion matrix.
    
    

```python
from sklearn.model_selection import train_test_split
from sklearn import neighbors
from imblearn.over_sampling import SMOTE
from sklearn.metrics import confusion_matrix

# Construction of the feature (X) and target (y) data
X = df.drop(['quality','grade'], axis=1).values
y = df['grade'].values

# Split the data to get a training se and a test set
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.1, random_state=1)

# knn Classifier is usd to train our ML 
kNN= neighbors.KNeighborsClassifier(n_neighbors=9)

# train the clasfier
kNN.fit(X_train, y_train)
print (kNN.score(X_test, y_test))

# Print confusion matrix 
y_pred= kNN.predict(X_test)
cf_mat= confusion_matrix(y_true=y_test, y_pred=y_pred)

print('Confusion matrix:\n', cf_mat)


# Use SMOTE to rebalance the dataset
sm= SMOTE(random_state=0)
X_train, y_train= sm.fit_sample(X_train, y_train)

# Train the classifier 
kNN.fit(X_train, y_train)

print (kNN.score(X_test, y_test))

# Print confusion matrix 
y_pred= kNN.predict(X_test)
cf_mat= confusion_matrix(y_true=y_test, y_pred=y_pred)
print('Confusion matrix:\n', cf_mat)
```

    
    
    Despite the reduction in term of accuracy, we get a better precision for classifying the wine.
    It is then better to apply the SMOTE technique and then improve the accuracy by applying various ML technique for training.

   ### 2.1.6 Data processing summary
    
    On the basis of the study which has been carried above, we can propose a function which processes the data as described above:
    The data will be scaled by the Normalizing process or the Standardizing process.
    Then, the outliers will be optionally removed.
    
    

```python

## Sugested methodology to preprocess the data
def Data_Preprocess(df, scaling, outlier):
# Check if we normalize or standardize the data
    if scaling == 'std':
        scalingObj = preprocessing.StandardScaler()
    else:
        scalingObj = preprocessing.MinMaxScaler()

    newdf = df.copy()
    
    # Apply the scaling to the data
    newdf[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']] = scalingObj.fit_transform(newdf[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']])

    # Construction of the feature (newdf_X) and target (newdf_y) data
    newdf_X = newdf.drop(['quality','grade'], axis=1).values
    newdf_y = newdf['grade'].values

    # Remove the outliers if required
    if outlier == 'remove':
        newdf_X = np.delete(newdf_X,(1079,1081), axis=0)
        newdf_y = np.delete(newdf_y,(1079,1081), axis=0)

    # Rebalance the data by applying the SMOTE technique (oversampling the data)
    sm= SMOTE(random_state=0)
    newdf_X, newdf_y = sm.fit_sample(newdf_X, newdf_y)
    return (newdf_X, newdf_y)

```




## 2.2 Features Selection
    
    An important topic before training the ML is to define the features to be considered to train the ML.
    Indeed, keeping meaningless feature may reduce the accuracy of our algorithm as those meaningless features act as noises for our ML. The idea is then to apply techniques which highlights the meaningful features to be selected in order to train our machine.
    The other reason to process to the feature Selection is to reduce the dimensionality of the database. Large database are indeed very difficult to handle and draw conclusion. Therefore, selectiong few meaningful features could make sense for that situation. In our case, the dataset has a dimentionality of 11 (11 features). This is not a very big database; but it could make sens to check if all features are relevant to use for training our ML model.
    
    In ths part, we will explore 2 Features selection techniques:
        1) The tree-based Feature Selection : Random Forest
        2) The Greedy Feature Selection
        
        
        

   ### 2.2.1 Tree-based Feature Selection: Random Forest
    
    The Tree based feature selection builds many trees and calculates the average reduction in uncertainty achieved by each feature across all the trees and uses this as a means of feature ranking.
    The typical Tree-based feature seection is the RandomForest.
    By applying the Random Forest algorithm, we are able to see the importance of each feature respectively to the grade of the red wine.
    
    

```python

from sklearn.ensemble import RandomForestClassifier

# Construction of the feature (newdf_X) and target (newdf_y) data and preprocessing the data
newdf_X, newdf_y = Data_Preprocess(df, 'std', 'keep')

# random forest Classifier is used to train for our ML 
forest = RandomForestClassifier(n_estimators=800, random_state=0)
forest.fit(newdf_X, newdf_y)

# store the features importances values
importances= forest.feature_importances_ 

# display the score of the feature importance
for index in range(len(newdf_X[0])):
    print ("Importance of feature ", index, ", ", newdf.keys()[index], "is", importances[index])

features_importance = list(reversed(np.argsort(importances)))

print ('The list of the Features classified accordingly to their importance is :', features_importance)
```

 

    The issue in the Tree based feature selection is that we can see the importance of the feature; but we are unable to determine if the feature is necessary (to be kept) in order to train optomally the ML.
    Another technique is very apropriate for that: the Greedy Feature selection.
    
    


   ### 2.2.2 Greedy Feature Selection
    
    The Greedy Feature Selection automatically reduces the number of features involved in a learning model based on their effective contribution to the overall accuracy performance of the algorithm.
    More specifically, the RFECV class provides a greedy backward search that can provide information on the number of useful features and automatically transform the training data into a reduced variable set.
    By applying the RFECV, we are able to see the features which needs to be selected toimprove our ML.
    We will apply successively the RFECV to dataset without outliers and with outliers and see the impact with the kNN and random Forest classifier.
    
    

    Different study on RFECV for diffierent classifier K Nearest Neighbors and Random Forest with and wthout outlier has been done. the result of the RFECV technique is presented below 

    We will first test the RFECV technique on data with removed outliers

```python

### RFECV technique on kNN with removing the outliers

from sklearn.feature_selection import RFECV
from sklearn.neighbors import KNeighborsClassifier
from sklearn import linear_model
from sklearn import model_selection
import matplotlib.pyplot as plt


# Construction of the feature (newdf_X) and target (newdf_y) data and preprocess the data
newdf_X, newdf_y = Data_Preprocess(df, 'norm', 'remove')

# knn Classifier is used to train for our ML 
kNN= neighbors.KNeighborsClassifier(n_neighbors=1)
estimator = linear_model.LogisticRegression()

# display the crossvalidation accuracy score 
# using the k fold cross validation (here k=10)
scores = model_selection.cross_val_score(kNN, newdf_X, newdf_y, cv=10)
print ('Initial Result',scores.mean())

# Use the RFECV algo in the data
selector = RFECV(estimator, cv=10)
selector.fit(newdf_X, newdf_y)

# Print the optimal number of feature to be used for our ML model
nb_features = selector.n_features_
print ('Optimal number of features is ', nb_features)

for index in range(len(selector.ranking_)):
    print('Feature importance:', newdf.keys()[index], "score: ", selector.ranking_[index] )
# ranking of each feature
print (selector.ranking_)
print (selector.support_)


# Do a test : select only the recommended number of features ...
newdf_X = newdf_X[ : , selector.support_ ]

#... and do a cross vaidation
# using the k fold cross validation (here k=10)
scores = model_selection.cross_val_score(kNN, newdf_X , newdf_y, cv=10)

print ('Result after feature selection  ',scores.mean())

# Prompt a graphic showing the score depending on the number of features selected
plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Cross validation score (nbof correct classifications)")
plt.plot(range(1, len(selector.grid_scores_) + 1), selector.grid_scores_)
plt.show()
```

    For that we will divide the dataset into 2 tables: A table of good wines and a table of bad wines and make differential in term of their features to see which features are the most relevant for one category (good wine) to another one (bad wine).
    This part has been essentially reused by the work of Yongho Kim in 'Prediction of Red Wine Quality(93.215%)'

    

```python
#Set 2 target criteria one for good wine, the other for bad wine
good = df[df.grade == 1]
notgood = df[df.grade == 0]

# Create 2 dataframe table for good and bad wine and calculate the mean of each feature
drop_items = ['quality','grade']
g1 = pd.DataFrame(good.drop(drop_items, axis=1).mean(), columns=['Good']).T
g2 = pd.DataFrame(notgood.drop(drop_items, axis=1).mean(), columns=['Not Good']).T

# calculate the mean of each feature for the complete list of wine
total = pd.DataFrame(df.drop(drop_items, axis=1).mean(), columns=['Total Average']).T

# Create a list for the 3 above sets of values (mean of each features)
data = g1.append([g2, total])

# Set standard
temp1 = data.values.reshape((3, 11))
standard = data.loc['Total Average'].values.reshape((1, 11))
temp = 100* temp1 / standard

# Express the values of data in percentage
data_percentage = pd.DataFrame(temp, columns = data.columns.values.tolist())

# For each set of values in date (sets of good wine, bad wine and total) stores the values for each anglr of pie chart
from math import pi
Attributes =list(data_percentage)
AttNo = len(Attributes)

values = data_percentage.iloc[0].tolist() #
values += values [:1]

angles = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles += angles [:1]

values2 = data_percentage.iloc[1].tolist() # 
values2 += values2 [:1]

angles2 = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles2 += angles2 [:1]

values3 = data_percentage.iloc[2].tolist() # 
values3 += values3 [:1]

angles3 = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles3 += angles3 [:1]

# Plot the figure
plt.figure(figsize=(10,10))
ax = plt.subplot(111, polar=True)
plt.xticks(angles[:-1],Attributes)

# Good 
ax.plot(angles, values, color = 'g')
ax.fill(angles, values, 'green', alpha=0.1)

# Not Good
ax.plot(angles2, values2, color = 'r')
ax.fill(angles2, values2, 'red', alpha=0.1)

# Total Average
ax.plot(angles3, values3, color = 'black')
ax.fill(angles3, values3, 'black', alpha=0.1)

plt.figtext(0.4,0.82,'Good Quality Average',color='green')
plt.figtext(0.28,0.48,'Not Good Quality Average',color='red')
plt.figtext(0.36,0.23,'Total Average',color='black')
plt.show()

```

## 3.3 Selected Machine Learning Model
    
    In the above section, we have tried to optimize the hyper parameters for Random Forest Classifier Model and the K Nearest Neighbors Classifiers. The results for both of them are almost same. But the KNN classifer has the advantage to be more stable than the Random Forest and has a better chance to be parametrable. Our selected Machine Learning will therefore be the KNN Classififer.
    
    To summarize, here are the operations which have been taken for our selected ML model.
    _ the wine database is read
    _ A new target result is created to define a good wine (wine >= quality 7)
    _ The data is Normalized 
    _ Outliers are kept
    _ To balance the data the SMOTE algorithm is applied
    _ 8 features are selected to train our model : fixed acidity, volatile acidity, residual sugar, free sulfur dioxide, total sulfur dioxide, density, sulphates, alcohol
    _ Then the KNN classfier is launched with the following parameter: algorithm='auto', n_neighbors=2, p=2, weights='uniform'


    

```python
import numpy as np
import pandas as pd
import os
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
from imblearn.over_sampling import SMOTE

# Function proposal for our ML model based on knn classfier
def ML_wine_quality (file, x_test):

    # read csv file and put it in data frame format
    df = pd.read_csv(file)
    
    # define new grade for the wine
    df['grade'] = 1 # good
    df.grade[df.quality < 7] = 0 # not good

    # Scaling data by normalizing features values
    scalingObj = preprocessing.MinMaxScaler()
    newdf = df.copy()
    newdf[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']] = scalingObj.fit_transform(newdf[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']])

    # Construction of the feature (newdf_X) and target (newdf_y) data 
    newdf_X = newdf.drop(['quality','grade'], axis=1).values
    newdf_y = newdf['grade'].values

    # Aplly the SMOTE technique to oversample data and rebalance the data
    sm= SMOTE(random_state=0)
    newdf_X, newdf_y = sm.fit_sample(newdf_X, newdf_y)
    
    # Elaborate the features to be selected for our model
    feature_selected = np.array([ True,  True, False,  True,  True, False,  True,  True, False,  True,  True])
    
    # Construct features data on only features slected
    newdf_X = newdf_X[ : , feature_selected ]
    
    # Train the model with knn classifier n_neighbors =2 and p=2
    KN = KNeighborsClassifier(algorithm='auto', n_neighbors=2,  p=2, weights='uniform')
    KN.fit(newdf_X, newdf_y)
    
    # do a prediction on the test data
    KN_pred = KN.predict(x_test)

    return (KN_pred)
```

## 4.2 Learning Curve
    
    Another way to evaluate the model is to analyze the leaning curve. 
    The learning curve gives basically the score and the gap between the training score and the cross validation score.
    Those information are useful to understand if the the model has a high varance (which means that the model is overfitted and applies very well on training set but not on other test set) or has a high-bias (which means that the model is underfited and cannot apply correctly to any training or test set). 
    Ideally, the 2 curves should join and have a good score.

```python
#### Learning curve
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import learning_curve
from sklearn.model_selection import ShuffleSplit

newdf_X, newdf_y = Data_Preprocess(df, 'norm', 'keep')
newdf_X = newdf_X[ : , feature_selection1 ]


# Cross validation with 100 iterations to get smoother mean test and train
# score curves, each time with 20% data randomly selected as a validation set.

custom_cv= ShuffleSplit(n_splits=100, test_size=0.2, random_state=0)

estimator = KNeighborsClassifier(algorithm='auto', n_neighbors=2,  p=2, weights='uniform')

title = "Learning Curve Wine quality"
plt.title(title)
ylim=(0.0, 1.01)

plt.ylim(ylim)
plt.xlabel("Training examples")
plt.ylabel("Score")

train_sizes=np.linspace(.05, 1.0, 10)

train_sizes, train_scores, test_scores= learning_curve(estimator, newdf_X, newdf_y, train_sizes=train_sizes, cv = custom_cv)
train_scores_mean= np.mean(train_scores, axis=1)
test_scores_mean= np.mean(test_scores, axis=1)

plt.grid()

plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-validation score")

plt.legend(loc="best")
plt.show()


```
