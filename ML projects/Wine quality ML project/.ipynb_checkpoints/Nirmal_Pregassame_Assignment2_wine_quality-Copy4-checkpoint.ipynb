# Q4
The operation of Forward chaining consist on :
1) Fire/remove any rule whose premises are satisfied in the
Knowledge Base
2) Add those rules to the Knowledge Base
3) Then  Repeat the above 2 operations until a conclusion can be given on the final sentence

We have though to assume that the KB is of Horn clauses.
The forward chaining is data driven


The operation of Backward chaining is the reverse operation in comparison to forward chaining.
The final atomic sentence is known and proven and the operation consiste on recursively search for the premises which proves the final sentence.
The backward chaining is goal driven.
and just like the forward chaining, we have to assume that the KB is of Horn clauses.

The main disadvantages of Forward chaining is that at a given step, the choice between the firable rules is random. Therefore, it may apply rules that are useless for the proof of the final sentence.

The main disadvantage of backward chaining is that it follows a particular path of reasoning even if it proves that it may be pointless.

```python
bins = (1, 6.5, 12)
group_names = ['bad', 'good']
df['quality'] = pd.cut(df['quality'], bins=bins, labels=group_names)
df['quality'] = df['quality'].cat.codes
X = df.drop(['quality'], axis=1)
y = df[['quality']]
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2, random_state=1511)
#SVC generates an accuracy of around 0.91

t0 = time()
clf = SVC(C=1, kernel='rbf')
clf.fit(train_X, train_y)
print('Training time', round(time() - t0, 3), 's')
pred = clf.predict(test_X)
pred = pred.tolist()
for i in range(len(pred)):
  pred[i] = round(pred[i])
print('Score', accuracy_score(test_y, pred))
```
